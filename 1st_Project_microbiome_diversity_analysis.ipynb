STEP 2: LOAD DATA SAFELY 

```python
import os
import requests

os.makedirs("data", exist_ok=True)

biom_url = "https://raw.githubusercontent.com/biocore/scikit-bio/main/skbio/util/tests/data/feature-table.biom"
meta_url = "https://raw.githubusercontent.com/biocore/scikit-bio/main/skbio/util/tests/data/sample_metadata.tsv"

# Download BIOM
r = requests.get(biom_url)
print("BIOM status:", r.status_code)

with open("data/feature-table.biom", "wb") as f:
    f.write(r.content)

# Download metadata
r2 = requests.get(meta_url)
print("Metadata status:", r2.status_code)

with open("data/sample_metadata.tsv", "wb") as f:
    f.write(r2.content)

print("Download complete")
```
